<h2><b>Question: How to calculate summary statistics?</b></h2>
Answer: - Basic statistics (mean, median, min, max, counts…) are easily calculable.<br>
- These or custom aggregations can be applied on the entire data set, a sliding window of the data, or grouped by categories.<br>
- The latter is also known as the split-apply-combine approach.<br>

<h2><b>Remember</b></h2>
- Aggregation statistics(mean, median, min, max, counts…) can be calculated on entire columns or rows.<br>
- groupby provides the power of the split-apply-combine pattern.<br>
- value_counts is a convenient shortcut to count the number of entries in each category of a variable.<br>

```python
discrete_df = df.select_dtypes(include=['object'])

numerical_df = df.select_dtypes(include=['int64', 'float64'])
```

```python
def discrete_univariate_analysis(discrete_data):
    for col_name in discrete_data:
        print("*"*10, col_name, "*"*10)
        print(discrete_data[col_name].agg(['count', 'nunique', 'unique']))
        print('Value Counts: \n', discrete_data[col_name].value_counts())
        print()
```

```python
discrete_univariate_analysis(discrete_df)
```

```python
def numerical_univariate_analysis(numerical_data):
    for col_name in numerical_data:
        print("*"*10, col_name, "*"*10)
        print(numerical_data[col_name].agg(['min', 'max', 'mean', 'median', 'std']))
        print()
```

```python
numerical_univariate_analysis(numerical_df)
```

```python
# List of numerical columns to analyze
numerical_cols = ['10percentage', '12percentage', 'collegeGPA', 'conscientiousness', 
                  'agreeableness', 'extraversion', 'neuroticism', 'openness_to_experience']
```

```python
# Create subplots for each numerical column
fig, axes = plt.subplots(4, 2, figsize=(14, 16))
fig.suptitle('Distribution of Numerical Columns', fontsize=16)

# Iterate over each numerical column and create histograms
for ax, col in zip(axes.flatten(), numerical_cols):
    sns.histplot(df[col], bins=30, kde=True, ax=ax, color='lightblue')
    ax.set_title(f'Distribution of {col}')

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
# Boxplot for outlier analysis of numerical columns
plt.figure(figsize=(14, 8))
sns.boxplot(data=df[numerical_cols], palette="Set2")
plt.xticks(rotation=45)
plt.title('Boxplot Analysis of Numerical Columns')
plt.show()
```

```python
# Scatter plots for relationships between numerical columns
sns.pairplot(df[numerical_cols])
plt.suptitle('Pairplot of Numerical Columns', y=1.02)
plt.show()
```

```python
# Countplots for categorical columns
categorical_columns = ['Designation', 'JobCity', 'Gender', '10board', '12board', 'Degree', 'Specialization', 'CollegeState']
```

```python
# Research Question Analysis
# Times of India claim: Fresh CS graduates can earn up to 2.5-3 lakhs in certain roles

# Filter for relevant designations and recent graduates
relevant_roles = ['Programming Analyst', 'Software Engineer', 'Hardware Engineer', 'Associate Engineer']
recent_graduates = df[df['Designation'].isin(relevant_roles) & (df['GraduationYear'] >= df['GraduationYear'].max() - 1)]

# Calculate average salary for recent graduates in relevant roles
avg_salary = recent_graduates['Salary'].mean() / 100000  # Convert to lakhs

print(f"Average salary for recent CS graduates in relevant roles: {avg_salary:.2f} lakhs")

# Plot salary distribution for recent graduates in relevant roles
plt.figure(figsize=(10, 6))
sns.histplot(recent_graduates['Salary'] / 100000, kde=True)
plt.axvline(x=2.5, color='r', linestyle='--', label='2.5 lakhs')
plt.axvline(x=3, color='g', linestyle='--', label='3 lakhs')
plt.title('Salary Distribution for Recent CS Graduates in Relevant Roles')
plt.xlabel('Salary (in lakhs)')
plt.ylabel('Frequency')
plt.legend()
plt.show()

# Calculate percentage of recent graduates earning within the claimed range
within_range = ((recent_graduates['Salary'] >= 250000) & (recent_graduates['Salary'] <= 300000)).mean() * 100

print(f"Percentage of recent graduates earning between 2.5-3 lakhs: {within_range:.2f}%")

# Relationship between college tier and salary
plt.figure(figsize=(10, 6))
sns.boxplot(x='CollegeTier', y='Salary', data=df)
plt.title('Relationship between College Tier and Salary')
plt.xlabel('College Tier')
plt.ylabel('Salary')
plt.show()

# Correlation between GPA and salary
correlation = df['collegeGPA'].corr(df['Salary'])
print(f"Correlation between college GPA and Salary: {correlation:.2f}")

plt.figure(figsize=(10, 6))
sns.scatterplot(x='collegeGPA', y='Salary', data=df)
plt.title('Relationship between College GPA and Salary')
plt.xlabel('College GPA')
plt.ylabel('Salary')
plt.show()
```

```python
# Select numerical columns for correlation analysis
numerical_data = df[['Salary', '10percentage', '12percentage', 'collegeGPA', 
                       'conscientiousness', 'agreeableness', 'extraversion', 
                       'neuroticism', 'openness_to_experience']]

# Calculate the correlation matrix
corr_matrix = numerical_data.corr()

# Create a heatmap for the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, linewidths=0.5)
plt.title('Correlation Heatmap of Numerical Variables')
plt.show()
```
